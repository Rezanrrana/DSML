```python
# Set operations
set1={5,6,4,3,9,1}
set2={1,12,4,7,3,2}


# Q1. find AuB

result = set1.union(set2)
print(result)
```

```python
#Q2. Fins the intersection

result = set1&set2
print(result)
```

```python
result = set1-set2
print(result)
```

```python
set1 ^ set2 # symmetric difference
```

```python
(set1-set2).union(set2-set1) # symmetric difference
```
