```python
# Break statement: it is use to break the loop from where we want

count = 1
while count < 10:
  if count == 3:
    break
  print (count)
  count += 1
print("Loop ended ")

```

```python
# Continue statement: it is use to scape the provided number from loop

count = 1
while count < 10:
 count += 1
 if count == 5:
  continue
 print(count)

print("Loop ended")

```

```python
# Pass

count = 1
while count < 10:
 count += 1
 if count == 5:
  pass
 print(count)

print("Loop ended")
```

```python
# class task 2 atm

print(" Enter the card first")

balance = 200

while True:
  print("-- ATM Menu--")
  print("Choose the option")

  print("1: to check balance")
  print("2: to deposit money")
  print("3: to exit")

# Check Balance
  option = int(input("Enter the option: "))

  if option == 1:
    print("Your balance is: ", balance)

# Deposit Money
  elif option == 2:
    amount = int(input("Enter the amount: "))
    balance += amount
    print("Your new balance is: ", balance)

# Exit
  elif option == 3:
    print("Exit")
    break
  else:
    print("Invalid option")


```
