1. Write a function toatal_salary(*args) that accepts an employee's salary components (basic salary, allowance, bonus etc.) return the total salary.

```python
def toatal_salary(*args):
    """
    Accepts any number of salary components (basic salary, allowance, bonus, etc.)
    and returns their total.
    """
    return sum(args)


# Example usage

basic = 30000
allowance = 5000
bonus = 2000

total = toatal_salary(basic, allowance, bonus)
print("Total Salary:", total)


# Works with any number of components
print(toatal_salary(20000, 1500))
print(toatal_salary(15000, 2000, 1000, 500))
```

2. Create a function cart_total(*args) that accept the price of any number of products and returns the total bill amount

```python
def cart_total(*args):
    """
    Accepts the price of any number of products and returns the total bill amount.
    """
    return sum(args)

rice = 500
milk =300
bread = 100
sugar =120

total = cart_total(rice, milk, bread, sugar)
print("Total Bill Amount:", total)
```

3. Write a function student_details(**kwargs) that accepts a student's information as keyword arguments and display all the details.a


```python
def student_detail(**kwargs):
    """
    Accepts a student's information as keyword arguments and displays all the details.
    """
    print("Student_details")
    for key, value in kwargs.items():
        print(f"{key}: {value}")

student_detail (name = "Rezan",cast = "Rana", age = 20, city = "Karachi")
```

4. Creat a function product_info(**kwargs) that accepts product details such as name, price, brand and cateegory, then display them in readable format.

```python
def product_info(**kwargs):
    """
    Accepts product details as keyword arguments and displays them in a readable format.
    """
    print("=" * 30)
    print("       PRODUCT INFO")
    print("=" * 30)
    for key, value in kwargs.items():
        print(f"{key.capitalize():<10}: {value}")

product_info(name = "Aquarium", price = 20000, brand = "LV", category = "Home Decor")
```
