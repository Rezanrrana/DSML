**SHALLOW AND DEEP COPY**

```python
#SHALLOW COPY

lst = [[1,2,3,4],[5,6,7,8]]
lst1 = lst
print(lst)
print(lst1)
```

```python
lst2 =lst1.copy()
lst2
```

```python
lst3 =[[1,2,3],[4,5,6]]
lst4 =lst3.copy()
```

```python
lst3[1][1] =100
print("orginal")
print(lst3)
print("copy")
print(lst4)
```

```python
# DEEP COPY

import copy
lst5 =[[1,2,3],[4,5,6]]
lst6 =copy.deepcopy(lst5)
lst5[1][2]= 100
```

```python
print("orginal")
print(lst5)
print("copy")
print(lst6)
```
